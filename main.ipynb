Import pandas to read and manipulate data

```python
import pandas as pd
```

```python
data = pd.read_csv("Data/StudentPerformanceFactors.csv")
data
```

Lets start Preprocessing

```python
import sklearn
# Split data 70-30
# Calculate the number of rows corresponding to 70%
num_rows_70_percent = int(len(data) * 0.7)

# Get the first 70% of the DataFrame
train = data.iloc[:num_rows_70_percent]
test = data.iloc[:-num_rows_70_percent]
train
```

```python
from sklearn import preprocessing

data_preprocessed = train

# Set target value
target_train = train["Exam_Score"]

# Drop target from data_pre so we dont have it in training

data_preprocessed = data_preprocessed.drop(columns="Exam_Score")

# Encode to numeric values
encoder = preprocessing.OneHotEncoder()
encoded = pd.DataFrame(encoder.fit_transform(data_preprocessed[['Parental_Involvement', 'Access_to_Resources', 'Extracurricular_Activities', 'Motivation_Level', 'Internet_Access', 'Family_Income', 'Teacher_Quality', 'School_Type', 'Peer_Influence', 'Learning_Disabilities', 'Parental_Education_Level', 'Distance_from_Home', 'Gender']]).toarray(), columns=encoder.get_feature_names_out(['Parental_Involvement', 'Access_to_Resources', 'Extracurricular_Activities', 'Motivation_Level', 'Internet_Access', 'Family_Income', 'Teacher_Quality', 'School_Type', 'Peer_Influence', 'Learning_Disabilities', 'Parental_Education_Level', 'Distance_from_Home', 'Gender']))
encoded

data_preprocessed = data_preprocessed.drop(columns=['Parental_Involvement', 'Access_to_Resources', 'Extracurricular_Activities', 'Motivation_Level', 'Internet_Access', 'Family_Income', 'Teacher_Quality', 'School_Type', 'Peer_Influence', 'Learning_Disabilities', 'Parental_Education_Level', 'Distance_from_Home', 'Gender'])
data_preprocessed = data_preprocessed.join(encoded)
# data_preprocessed = pd.merge(data_preprocessed, encoded, on='index_column')
#REPLACING NAN WITH 0. NOT SURE WHY THERE ARE NaN
#data_preprocessed.iloc[:, 6:] = data_preprocessed.iloc[:, 6:].fillna(0)
data_preprocessed

```

Lets try and freestyle a prediction


```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(data_preprocessed, target_train)
```

Preprocess test data

```python
test_target = test["Exam_Score"]
test = test.drop(columns="Exam_Score")

# Encode to numeric values
encoder = preprocessing.OneHotEncoder()
encoded = pd.DataFrame(encoder.fit_transform(test[['Parental_Involvement', 'Access_to_Resources', 'Extracurricular_Activities', 'Motivation_Level', 'Internet_Access', 'Family_Income', 'Teacher_Quality', 'School_Type', 'Peer_Influence', 'Learning_Disabilities', 'Parental_Education_Level', 'Distance_from_Home', 'Gender']]).toarray(), columns=encoder.get_feature_names_out(['Parental_Involvement', 'Access_to_Resources', 'Extracurricular_Activities', 'Motivation_Level', 'Internet_Access', 'Family_Income', 'Teacher_Quality', 'School_Type', 'Peer_Influence', 'Learning_Disabilities', 'Parental_Education_Level', 'Distance_from_Home', 'Gender']))
encoded

test = test.drop(columns=['Parental_Involvement', 'Access_to_Resources', 'Extracurricular_Activities', 'Motivation_Level', 'Internet_Access', 'Family_Income', 'Teacher_Quality', 'School_Type', 'Peer_Influence', 'Learning_Disabilities', 'Parental_Education_Level', 'Distance_from_Home', 'Gender'])
test = test.join(encoded)
# REPLACING NAN WITH 0. NOT SURE WHY THERE ARE NaN
test
```

Predict on test


```python
prediction = lr.predict(test)

# add the prediction and gold standard (target) to the dataframe
school_comparison = test.assign(Exam_Score=test_target)
school_comparison = school_comparison.assign(predictionSchool=prediction)

school_comparison
```

Evaluation


```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(test_target, prediction)
mae
```
